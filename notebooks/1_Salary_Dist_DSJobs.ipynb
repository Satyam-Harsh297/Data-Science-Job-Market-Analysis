```python
import pandas as pd
import matplotlib.pyplot as plt
from datasets import load_dataset
import ast
import seaborn as sns


dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df['job_month'] = pd.to_datetime(df['job_posted_date']).dt.strftime('%b')
df =  df.dropna(subset = ['job_skills'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x))
```

```python
# Filtering for Country and Jobs

df1 = df[(df['job_title_short'].str.contains('Data')) & (df['job_country'] == 'United States')]

df1
```

```python
lis = [x for x in df1['job_title_short'].unique()]
print(lis)
```

```python
fig, ax = plt.subplots(figsize=(12, 7))
job_order = df1.groupby('job_title_short')['salary_year_avg'].median().sort_values(ascending=False).index
sns.boxplot(data = df1 , x = 'salary_year_avg' , y = 'job_title_short' , order = job_order, ax=ax)
ax.set_xlim(0,600000)
ax.set_ylabel("")
ax.set_xlabel("Annual Salary" ,fontsize = 12)
ax.set_title("Salary Distribution of Data Science Jobs\nin the United States" , fontsize = 18)
```

```python
import matplotlib.ticker as mticker

plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['font.sans-serif'] = ['DejaVu Sans']

job_order = df1.groupby('job_title_short')['salary_year_avg'].median().sort_values(ascending=False).index

fig, ax = plt.subplots(figsize=(12, 7))

sns.boxplot(
    data=df1,
    x='salary_year_avg',
    y='job_title_short',
    order=job_order,
    palette=['#4C72B0', '#55A868', '#C44E52', '#8172B2', '#CCB974', '#64B5CD'],
    width=0.5,
    linewidth=1.2,
    flierprops=dict(marker='o', markerfacecolor='gray', markersize=4, alpha=0.5),
    ax=ax
)

ax.set_title('Salary Distribution of Data Science Jobs\nin the United States', 
             fontsize=18, fontweight='bold', loc='left', pad=15)
ax.set_xlabel('Annual Salary', fontsize=11)
ax.set_ylabel('')
ax.xaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'${x/1000:,.0f}K'))

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_color('#cccccc')
ax.spines['bottom'].set_color('#cccccc')

ax.grid(axis='x', linestyle='--', alpha=0.4, color='gray')
ax.set_axisbelow(True)
ax.tick_params(axis='both', labelsize=10)
ax.set_xlim(0, 600000)
plt.tight_layout()
plt.show()
```
