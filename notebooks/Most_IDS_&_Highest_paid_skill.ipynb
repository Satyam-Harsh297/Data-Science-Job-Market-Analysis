```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt
import ast

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df['job_month'] = pd.to_datetime(df.job_posted_date).dt.strftime("%b")
df = df.dropna(subset=['job_skills'])
df['job_skills'] = df['job_skills'].apply(lambda skill: ast.literal_eval(skill))
```

```python
df
```

```python
df_exploded = df.explode('job_skills')
```

```python
df_exploded
```

```python
df_exploded = df_exploded[(df_exploded['job_country'] == 'United States') & (df_exploded['job_title_short'] == 'Data Analyst')].copy()
```

```python


df_exploded
```

```python
Top_ten_MDS = list(df_exploded.job_skills.value_counts().head(10).index)
df_exploded1 = df_exploded[df_exploded['job_skills'].isin(Top_ten_MDS)].copy()
Top_ten_MDS_df = df_exploded1.groupby('job_skills').agg(median_salary = ('salary_year_avg' , 'median'))
Top_ten_MDS_df = Top_ten_MDS_df.sort_values('median_salary' , ascending=False)
Top_ten_MDS_df

```

```python
Highest_paid_skills = df_exploded.groupby('job_skills').agg(median_salary = ('salary_year_avg' , 'median'))
Highest_paid_skills = Highest_paid_skills.sort_values('median_salary' , ascending=False).head(10)
Highest_paid_skills
```
